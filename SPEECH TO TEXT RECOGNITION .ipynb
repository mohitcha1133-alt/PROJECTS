```python
# üéôÔ∏è Speech-to-Text System using SpeechRecognition
# =================================================
# This notebook converts short audio clips to text using the Google Web Speech API.

# Step 1: Install dependencies
!pip install SpeechRecognition pydub --quiet
!sudo apt-get install ffmpeg -y --quiet

# Step 2: Import libraries
import speech_recognition as sr
from IPython.display import Audio, display

# Step 3: Upload audio file (WAV/MP3)
from google.colab import files
uploaded = files.upload()

filename = list(uploaded.keys())[0]
print(f"\nUploaded file: {filename}")
display(Audio(filename))

# Step 4: Initialize recognizer and transcribe
def transcribe_audio(file_path):
    recognizer = sr.Recognizer()
    with sr.AudioFile(file_path) as source:
        print("\nListening to audio...")
        audio = recognizer.record(source)

    try:
        print("Transcribing...")
        text = recognizer.recognize_google(audio)
        print("\n‚úÖ Transcription Complete:")
        print("------------------------")
        print(text)
        return text
    except sr.UnknownValueError:
        print("‚ùå Could not understand the audio.")
    except sr.RequestError as e:
        print(f"‚ö†Ô∏è API request error: {e}")

# Step 5: Run transcription
transcribed_text = transcribe_audio(filename)

# Step 6 (Optional): Save transcript
if transcribed_text:
    with open("transcription.txt", "w") as f:
        f.write(transcribed_text)
    print("\nüíæ Transcript saved as 'transcription.txt'")
```
