```python
# üìò Article Summarizer using NLP (Hugging Face Transformers)

from transformers import pipeline

def summarize_article(text, max_length=150, min_length=40):
    """
    Summarize long text using a pre-trained transformer model.
    :param text: Input text (article/content)
    :param max_length: Maximum length of the summary
    :param min_length: Minimum length of the summary
    :return: Summarized text
    """
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)
    return summary[0]['summary_text']


if __name__ == "__main__":
    # Example Input Article
    article = """
    Artificial intelligence (AI) is transforming the world in unprecedented ways.
    From healthcare to finance, AI-driven technologies are enhancing productivity,
    improving decision-making, and creating innovative solutions to complex problems.
    However, as AI becomes more integrated into society, ethical concerns around data
    privacy, bias, and job displacement continue to grow. Policymakers, researchers,
    and organizations are now working together to establish frameworks that ensure
    responsible and transparent use of AI systems.
    """

    print("üìÑ Original Article:\n", article)
    print("\nüîç Generating Summary...\n")

    summary = summarize_article(article)
    print("üìù Summary:\n", summary)
```
